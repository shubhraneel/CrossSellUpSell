```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 50)
```

```python
data = pd.read_csv("Data.csv")
data
```

```python


dfT = pd.DataFrame()

for W in data["Ship-to nu"].unique():
    
    d = data[data["Ship-to nu"]==W]
    d['Doc. Date'].unique()

    # frequency={}
    # for i in d["Material"]:
    #     if i in frequency:
    #         frequency[i]+=1
    #     else: frequency[i]=1

    trend ={}
    for M in d["Material"].unique():
        u=[]
        for i in sorted(d['Doc. Date'].unique()):
            x = (d[d["Doc. Date"]==i])
            
            if M in x['Material'].unique(): u.append(1)
            else: u.append(0)

        print(u); trend[str(M)]=[u]
        
    df = pd.DataFrame(trend).T
    df = df.reset_index()
    df["Wholesaler"] = [W]*len(df)
    dfT = pd.concat([dfT, df], axis=0)
```

```python
dfT.columns = ["Material", "Trend", "Wholesaler"]
dfT = dfT[["Wholesaler", "Material", "Trend"]]
dfT = dfT.reset_index(drop=True)
# dfT.to_csv("trend_user_material_all.csv", index=False) #uncomment to save user_material_trend_file
```

```python
def remove_preeciding_zero(l):
    for i in range(len(l)):
        if l[i]==1: return l[i:]
    else: return l

def count1(l):
    return (l.count(1)/len(l))*100

def cont_z(l):
#     for i in range(len(lp)):
#         if lp[i]==1: start=i; break
            
#     l=lp[start:]
    flag=0; c=0; j=[]
    for i in range(len(l)):
        if l[i]==1 and flag==0:j.append(0); continue
        if l[i]==0 and flag==0: flag=1;
        
        if l[i]==1 and flag==1: flag=0; j.append(c); c=0
        if flag==1: c+=1;
        if i==len(l)-1 and flag==1: flag=0; j.append(c)
        
        
    ser = (j[1:])
    if len(ser)==0: return 0.5
    ans = np.mean(ser)
    
    if ans==0:
        if len(ser)<2: return 0.5
        else: return np.mean(ser+[1])
    
    else: return ans
```

```python
dfT["Trend"] = dfT["Trend"].apply(remove_preeciding_zero)

## total percent of oreders in which included - frequency
#orders made between two consecutive inclusions - gap


dfT["gaps"] = dfT["Trend"].apply(cont_z).fillna(0)
dfT["freq"] = dfT["Trend"].apply(count1)
```

```python
dfT
```

```python
dfT["gaps"] = dfT["gaps"].replace(0, 1.) 
dfT["rating"] = dfT["freq"]/dfT["gaps"]
dfT
```
